```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Patch
```

```python
fig, ax = plt.subplots(figsize=(6,5))

# Define the mineral categories
x = ['Andalusites', 'Biotites', 'Oolitic grains']

# Define the number of images for each category in different splits from Zhu et al. (2020).
train = [43, 96, 47]  # Training set
val   = [5, 10, 5]    # Validation set
test  = [5, 11, 5]    # Test set

# Define the number of images in "Our dataset" for each category
our_train = [0, 154, 0]  # Training images from our dataset
our_val   = [0, 19, 0]   # Validation images from our dataset
our_test  = [0, 14, 0]   # Test images from our dataset

# Plot stacked bars for each category
ax.bar(x, train, color='steelblue', edgecolor='black', label='Train')  # Standard training data
ax.bar(x, our_train, bottom=train, edgecolor='black', color='steelblue', hatch='/')  # Our dataset training data
ax.bar(x, val, bottom=np.add(our_train, train), edgecolor='black', color='sandybrown')  # Standard validation data
ax.bar(x, our_val, bottom=np.add(np.add(our_train, train), val), edgecolor='black', color='sandybrown', hatch='/')  # Our dataset validation data
ax.bar(x, test, bottom=np.add(np.add(our_train, train), np.add(val, our_val)), edgecolor='black', color='lightgreen')  # Standard test data
ax.bar(x, our_test, bottom=np.add(np.add(np.add(our_train, train), val), np.add(our_val, test)), edgecolor='black', color='lightgreen', hatch='/')  # Our dataset test data

# Define custom legend elements
legend_elements = [Patch(facecolor='lightgreen', edgecolor='black', label='Test'),
                   Patch(facecolor='sandybrown', edgecolor='black', label='Validation'),
                   Patch(facecolor='steelblue', edgecolor='black', label='Train'),
                   Patch(facecolor='white', edgecolor='black', hatch='/', label='Our dataset')]

# Add legend to the plot
ax.legend(handles=legend_elements, loc='upper right')

# Label for the y-axis
ax.set_ylabel('Number of images')

# Optionally, set the y-axis to a logarithmic scale (currently commented out)
# ax.set_yscale('log')
```
